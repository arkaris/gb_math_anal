```python
def pol_to_dec(a, r):
    return r*np.cos(a), r*np.sin(a)

a = np.linspace(0, 2*np.pi, 100)
r = [1]*100

x, y = pol_to_dec(a, r)
plt.plot(x, y)
plt.xlim(-1.1, 1.1)
plt.ylim(-1.1, 1.1)
```

```python
def dec_to_pol(x, y):
    r = np.sqrt(np.square(x) + np.square(y))
    a = np.arccos(x/r)
    return a, r

x = np.linspace(-10, 10, 100)
y = [1] * 100
a, r = dec_to_pol(x, y)
plt.polar(a, r)
```
