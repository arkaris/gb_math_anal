```python
import pylab
from mpl_toolkits.mplot3d import Axes3D
```

```python
x = np.linspace(-5, 5, 11)
y = np.linspace(-5, 5, 11)
x, y = np.meshgrid(x, y)
z1 = - x + y + 2
z2 = - x + y + 0
fig = pylab.figure()
ax = Axes3D(fig)
ax.plot_wireframe(x, y, z1)
ax.plot_surface(x, y, z2)
```
